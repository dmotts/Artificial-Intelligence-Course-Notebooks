# NumPy: Introdução e Manipulação de Arrays
O NumPy é uma biblioteca fundamental para computação científica em Python. Ela fornece suporte eficiente para a criação, manipulação e cálculos em arrays multidimensionais. Vamos começar com alguns conceitos básicos e exemplos de uso:

## Importação
Antes de começarmos a utilizar o NumPy, precisamos importar a biblioteca. Podemos fazer isso usando a seguinte linha de código:

```python
import numpy as np
```

Agora estamos prontos para começar!

## Criação de Arrays
O NumPy oferece várias maneiras de criar arrays. Aqui estão alguns exemplos:

```python
# Criando um array a partir de uma lista
arr1 = np.array([1, 2, 3, 4, 5])
arr1
```

```python
# Criando um array com valores sequenciais
arr2 = np.arange(0, 10, 2)  # Valores de 0 a 10 (exclusivo) com passo de 2
arr2
```

```python
# Criando um array de zeros
arr3 = np.zeros((3, 3))  # Array 3x3 preenchido com zeros
arr3
```

```python
# Criando um array de uns
arr4 = np.ones((2, 4))  # Array 2x4 preenchido com uns
arr4
```

```python
# Criando um array com valores aleatórios
arr5 = np.random.rand(5)  # Array de 5 elementos com valores aleatórios entre 0 e 1
arr5
```

```python
# Criando um array com valores espaçados
arr6 = np.linspace(0, 1, 6)  # Valores espaçados igualmente entre 0 e 1 com 6 elementos
arr6
```

## Manipulação de Arrays
O NumPy permite manipular arrays de várias maneiras. Aqui estão alguns exemplos:

```python
# Acessando elementos
print(arr1[0])  # Acessa o primeiro elemento do array
```

```python
# Acessando fatias
print(arr2[2:5])  # Acessa os elementos do índice 2 ao 4
```

```python
# Modificando elementos
arr3[1, 1] = 5  # Modifica o valor do elemento na posição (1, 1)
arr3
```

```python
# Operações com arrays
arr4 = arr1 + arr2  # Soma os elementos de arr1 e arr2 elemento por elemento
arr4
```

```python
# Reshape de arrays
arr5 = np.arange(10)
arr5
```

```python
arr6 = arr5.reshape((2, 5))  # Redimensiona o array para uma matriz 2x5
arr6
```

```python
# Operações matemáticas
arr7 = np.sin(arr5)  # Calcula o seno de cada elemento do array
arr7
```

```python
arr8 = np.exp(arr6)  # Calcula a exponencial de cada elemento da matriz
arr8
```

```python
# Funções estatísticas
mean = np.mean(arr5)  # Calcula a média dos elementos do array
mean
```

```python
std = np.std(arr6)  # Calcula o desvio padrão dos elementos da matriz
std
```

Neste exemplo, importamos o NumPy, criamos arrays usando diferentes métodos, manipulamos os arrays realizando operações, acessando elementos e fatias, modificando valores, realizando operações matemáticas e aplicando funções estatísticas.

# Pandas: Introdução e Manipulação de Dados Tabulares
O Pandas é uma biblioteca poderosa para análise e manipulação de dados em Python. Ele oferece estruturas de dados flexíveis e eficientes, como o DataFrame, que permite trabalhar com dados tabulares de forma fácil e intuitiva. Vamos começar com alguns conceitos básicos e exemplos de uso:

## Importação
Antes de começarmos a utilizar o Pandas, precisamos importar a biblioteca. Podemos fazer isso usando a seguinte linha de código:

```python
import pandas as pd
```

## Leitura de Dados
O Pandas oferece várias maneiras de ler dados de diferentes fontes. Aqui estão alguns exemplos:



```python
# Lendo um arquivo CSV
#df1 = pd.read_csv('dados.csv')

# Lendo um arquivo Excel
#df2 = pd.read_excel('dados.xlsx')

# Lendo dados de uma URL
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

## Manipulação de Dados
O Pandas oferece várias funcionalidades para manipular os dados em um DataFrame:

```python
# Selecionando colunas específicas
df[['Name', 'Sex']]
```

```python
# Filtrando dados com base em condições
df[df['Age'] > 10.0]
```

```python
# Ordenando o DataFrame
df.sort_values('Name', ascending=True)
```

```python
# Ordenando o DataFrame
df.sort_values('Name', ascending=False)
```

```python
# Agrupando dados e aplicando funções
df.groupby('Survived').mean()
```

```python
# Preenchendo valores ausentes
df.fillna('Cabin')
```
