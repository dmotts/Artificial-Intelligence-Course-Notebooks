# Naive Bayes
Naive Bayes é um algoritmo de aprendizado de máquina baseado no Teorema de Bayes, que é uma teoria estatística utilizada para calcular probabilidades condicionais. O algoritmo Naive Bayes é amplamente utilizado para problemas de classificação e é especialmente popular em tarefas de processamento de linguagem natural, como classificação de texto, análise de sentimentos e filtragem de spam.

A abordagem "Naive" (ingênuo) do algoritmo está relacionada à suposição de independência condicional entre as características (ou atributos) dos dados, ou seja, assume-se que cada característica é independente das outras, dado o rótulo da classe. Essa é uma suposição simplificada e por isso é chamada de "ingênua", pois, na realidade, nem sempre as características são realmente independentes. Apesar dessa simplificação, Naive Bayes muitas vezes apresenta bom desempenho em uma variedade de problemas da vida real.

O algoritmo Naive Bayes é baseado no Teorema de Bayes, que pode ser formulado da seguinte maneira:

```
P(C | X) = P(X | C) * P(C) / P(X)
```

onde:
- P(C | X) é a probabilidade da classe C dado o conjunto de características X.
- P(X | C) é a probabilidade de observar o conjunto de características X dado que a classe C é verdadeira.
- P(C) é a probabilidade a priori da classe C.
- P(X) é a probabilidade a priori do conjunto de características X.

Naive Bayes usa essa fórmula para calcular a probabilidade de cada classe para um novo conjunto de características e seleciona a classe com a maior probabilidade como a previsão do modelo.

Existem diferentes variantes do algoritmo Naive Bayes, dependendo da distribuição das características e dos tipos de dados. Alguns dos tipos comuns são:

- Naive Bayes Gaussiano: É usado para dados contínuos e assume que as características seguem uma distribuição normal (Gaussiana).
- Naive Bayes Multinomial: É usado para dados discretos, como contagens de palavras em um documento. É frequentemente aplicado em tarefas de classificação de texto.
- Naive Bayes Bernoulli: É uma variante do Naive Bayes Multinomial, mas é adequado para dados binários (por exemplo, representação de presença ou ausência de palavras em um documento).

O Naive Bayes é um algoritmo simples, rápido e eficaz em muitos cenários, especialmente em problemas de classificação com alta dimensionalidade. Embora a suposição de independência condicional nem sempre seja verdadeira nos dados do mundo real, o Naive Bayes ainda pode produzir resultados úteis e é especialmente adequado para conjuntos de dados menores ou problemas com restrições de recursos computacionais.

## Exemplo de Classificação com Naive Bayes usando o Conjunto de Dados Iris

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from mlxtend.plotting import plot_decision_regions
````

````python
#Preparação de dados
# Carregando o conjunto de dados "Iris"
data = load_iris()

# Criando um DataFrame com os dados
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Selecionando as duas características (comprimento da pétala e largura da pétala)
X = df[['petal length (cm)', 'petal width (cm)']]
y = df['target']

# Dividindo o conjunto de dados em treinamento e teste (70% treinamento, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando o objeto do modelo Naive Bayes Gaussiano
nb = GaussianNB()
````

````python
#Realização do treino
# Treinando o modelo com o conjunto de treinamento
nb.fit(X_train, y_train)
````

````python
# Realizando previsões no conjunto de teste
y_pred = nb.predict(X_test)

# Calculando a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia do modelo: {accuracy:.2f}')
````

````python
# Exibindo o relatório de classificação
print(classification_report(y_test, y_pred))
````

````python
# Plotando as regiões de decisão do modelo usando as duas características selecionadas
plot_decision_regions(X_train.values, y_train.values, clf=nb, legend=2)
plt.xlabel('Comprimento da Pétala (cm)')
plt.ylabel('Largura da Pétala (cm)')
plt.title('Naive Bayes - Regiões de Decisão')
plt.show()
````

## Exemplo de Classificação com Naive Bayes usando o Conjunto de Dados "Breast Cancer"

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from mlxtend.plotting import plot_decision_regions
````

````python
# Carregando o conjunto de dados "Breast Cancer"
data = load_breast_cancer()

# Criando um DataFrame com os dados
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Selecionando as duas características (mean radius e mean texture) para plotar as regiões de decisão
X = df[['mean radius', 'mean texture']]
y = df['target']

# Dividindo o conjunto de dados em treinamento e teste (70% treinamento, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando o objeto do modelo Naive Bayes Gaussiano
nb = GaussianNB()
````


No contexto do conjunto de dados de câncer de mama, que é frequentemente usado em tarefas de classificação para prever se um tumor é benigno ou maligno, os termos "mean radius" e "mean texture" referem-se a características específicas extraídas de imagens digitalizadas de aspirados de agulha fina (FNA) de tumores.

Mean Radius (Raio Médio): Este é um valor médio dos comprimentos médios das distâncias do centro até os pontos no contorno do tumor. Em outras palavras, é uma medida média das distâncias do centro do tumor até sua borda.

Mean Texture (Textura Média): Refere-se à média das variações de intensidade de tons de cinza na imagem, que pode ser associada à uniformidade da textura. Em termos mais simples, representa como os tons de cinza são distribuídos na imagem, proporcionando informações sobre a homogeneidade ou heterogeneidade da textura do tumor.

Essas características são apenas duas das várias medidas extraídas das imagens no conjunto de dados de câncer de mama. Elas são importantes porque podem fornecer informações valiosas para os modelos de aprendizado de máquina na tarefa de distinguir entre tumores benignos e malignos com base em características visuais extraídas das imagens médicas.


````python
# Treinando o modelo com o conjunto de treinamento
nb.fit(X_train, y_train)

# Realizando previsões no conjunto de teste
y_pred = nb.predict(X_test)

# Calculando a acurácia do modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia do modelo: {accuracy:.2f}')
````

````python
# Exibindo o relatório de classificação
print(classification_report(y_test, y_pred))
````

````python
# Plotando as regiões de decisão do modelo usando as duas características selecionadas
plot_decision_regions(X_train.values, y_train.values, clf=nb, legend=2)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Naive Bayes - Regiões de Decisão')
plt.show()
````

### Principal diferença do Naive Bayes, KNN e SVM
Os modelos Naive Bayes, k-Nearest Neighbors (KNN) e Support Vector Machines (SVM) são algoritmos de aprendizado de máquina que podem ser aplicados em tarefas de classificação. Aqui estão algumas das principais diferenças entre esses modelos:

Base Teórica:

Naive Bayes: Baseado no teorema de Bayes, assume independência condicional entre as variáveis preditoras.
KNN: Baseado na proximidade dos vizinhos. Classifica uma instância com base na classe predominante entre seus k vizinhos mais próximos.
SVM: Busca encontrar o hiperplano de decisão que melhor separa as classes no espaço de características.
Abordagem de Treinamento:

Naive Bayes: Usa estatísticas de probabilidade condicional para treinar o modelo.
KNN: Não requer um estágio de treinamento explícito; armazena os dados de treinamento e faz predições com base na proximidade das instâncias de teste aos pontos de treinamento.
SVM: Procura otimizar um hiperplano de decisão durante o treinamento.
Manuseio de Atributos:

Naive Bayes: Assume independência condicional entre os atributos, o que pode ser uma simplificação forte.
KNN: Lida bem com atributos irrelevantes ou redundantes, mas pode ser sensível à escala dos atributos.
SVM: Pode lidar efetivamente com conjuntos de dados de alta dimensionalidade e não é sensível à escala dos atributos.
Sensibilidade ao Ruído:

Naive Bayes: Pode ser sensível a atributos irrelevantes ou dependentes, mas geralmente é robusto.
KNN: Sensível ao ruído e pode ser afetado por outliers.
SVM: Geralmente robusto ao ruído e outliers, especialmente quando se utiliza um kernel apropriado.
Interpretabilidade:

Naive Bayes: Geralmente fácil de interpretar devido à sua simplicidade.
KNN: Menos interpretável, pois a decisão é baseada na proximidade com outros pontos.
SVM: Pode ser menos intuitivo de interpretar, especialmente em espaços de alta dimensão.
A escolha entre esses modelos depende das características específicas do problema, do tamanho do conjunto de dados, da dimensionalidade dos dados e de outras considerações relacionadas ao domínio do problema. Cada modelo tem suas vantagens e limitações, e a escolha deve ser feita com base nas características do seu conjunto de dados e nos objetivos da tarefa.
